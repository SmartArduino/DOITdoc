```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/home/scott/Dropbox/C.TheArtofWork/cool.csv')
```

```python
df.Diary = pd.to_datetime(df.Diary.values)
```

```python
df['ua'] = pd.to_datetime(df.ua.values).time
df['la'] = pd.to_datetime(df.la.values).time
df['lp'] = pd.to_datetime(df.lp.values).time
df['up'] = pd.to_datetime(df.up.values).time
```

```python
df = df.fillna(method='bfill')
df.head()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(12, 6))

ax[1].plot(df['Diary'], df['lp'], label='lp')
ax[1].plot(df['Diary'], df['up'], label='up')
ax[0].plot(df['Diary'], df['la'], label='la')
ax[0].plot(df['Diary'], df['ua'], label='ua')

labels = ax[0].get_xticklabels()
plt.setp(labels, rotation=30, fontsize=10)
labels = ax[1].get_xticklabels()
plt.setp(labels, rotation=30, fontsize=10)
fig.tight_layout()
```
